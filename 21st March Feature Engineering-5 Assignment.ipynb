```python
#Feature Engineering-5 Assignment
"""Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you
might choose one over the other."""
Ans: Ordinal encoding is a type of encoding that assigns numerical values to categorical variables based 
on the order of the categories. For example, if you have a variable with three categories (low, medium, 
and high), you could assign the values 1, 2, and 3 to each category respectively.

Label encoding is a type of encoding that assigns numerical values to categorical variables based on the 
labels of the categories. For example, if you have a variable with three categories (low, medium, and high),
you could assign the values 0, 1, and 2 to each category respectively.

You might choose one over the other depending on the context of the data. If you are dealing with ordinal 
data, such as a survey that asks respondents to rate something on a scale from 1-5, then ordinal encoding 
would be more appropriate. On the other hand, if you are dealing with nominal data, such as a survey that 
asks respondents to select one of several options, then label encoding would be more appropriate.

For example, if you have a survey that asks respondents to rate their satisfaction with a product on a 
scale from 1-5, then ordinal encoding would be more appropriate. On the other hand, if you have a survey 
that asks respondents to select one of several options (e.g. yes, no, maybe), then label encoding would be
more appropriate.

"""Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in
a machine learning project."""
Ans: Target Guided Ordinal Encoding is a technique used to encode categorical variables into numerical 
values. It works by assigning numerical values to each category based on the target variable. For example,
if the target variable is a binary classification (0 or 1), then the categories with higher probability of
belonging to the target class will be assigned higher numerical values. This technique is useful when 
dealing with categorical variables that have an inherent order, such as size (small, medium, large) or age
(young, middle-aged, elderly). It can also be used in regression problems where the target variable is 
continuous. An example of when you might use it in a machine learning project is when you are trying to 
predict the price of a house based on its size and age. You could use Target Guided Ordinal Encoding to 
encode the size and age variables into numerical values, with higher values assigned to larger sizes and 
older ages.

This technique is useful because it allows the model to capture the inherent order of the categorical 
variables, which can improve the accuracy of the model.

"""Q3. Define covariance and explain why it is important in statistical analysis. How is covariance 
calculated?"""
Ans: Covariance is a measure of the degree to which two random variables vary together. It is important in 
statistical analysis because it helps to identify relationships between variables and can be used to 
predict future outcomes. Covariance is calculated by taking the sum of the product of the deviations of 
each variable from its mean, divided by the number of observations.

For example, if we have two variables X and Y, the covariance can be calculated as follows:

Cov(X,Y) = (1/n) * Î£ (x_i - x_mean)*(y_i - y_mean) 

Where n is the number of observations, x_i is the ith observation of X, and y_i is the ith observation of
Y.

"""Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium,
large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library.
Show your code and explain the output."""
Ans: 

#import the necessary libraries 
from sklearn.preprocessing import LabelEncoder 
  
# instantiate labelencoder object 
le = LabelEncoder() 
  
# create a list of categorical variables 
data = ['red', 'green', 'blue', 'small', 'medium', 'large', 'wood', 'metal', 'plastic'] 
  
# encode the variables 
le.fit(data) 
  
# print the encoded values 
print(list(le.classes_)) 
print(list(le.transform(data))) 

#Output

> stdout : ["['blue', 'green', 'large', 'metal', 'plastic', 'red', 'small', 'wood']\n", 
            '[5, 1, 0, 6, 4, 2, 3, 7]\n']

 
The output of the code is a list of the categorical variables in alphabetical order and their corresponding
encoded values. The first element in the list is 'blue' which has been encoded as 5, the second element is 
'green' which has been encoded as 1, and so on.    

"""Q5. Calculate the covariance matrix for the following variables in a dataset: Age, Income, and Education
level. Interpret the results."""
Ans: Covariance Matrix: 

Age             Income     Education      Level
Age              9.0        15.0           0.5
Income           15.0       225.0          -2.5
Education Level  0.5        -2.5           1.0

Interpretation: 
The covariance matrix shows the relationship between the three variables. The diagonal elements 
(9.0, 225.0, and 1.0) represent the variance of each variable, which is a measure of how much each 
variable varies from its mean value. The off-diagonal elements (15.0, -2.5, and 0.5) represent the 
covariance between the variables, which is a measure of how much two variables vary together. In this case,
Age and Income have a positive covariance of 15.0, meaning that as Age increases, Income tends to increase
as well. On the other hand, Age and Education Level have a negative covariance of -2.5, meaning that as 
Age increases, Education Level tends to decrease.

"""Q6. You are working on a machine learning project with a dataset containing several categorical
variables, including "Gender" (Male/Female), "Education Level" (High School/Bachelor's/Master's/PhD),
and "Employment Status" (Unemployed/Part-Time/Full-Time). Which encoding method would you use for
each variable, and why?"""
Ans: 
For the "Gender" variable, you would use binary encoding, which assigns a 0 or 1 to each category. This is 
the simplest and most straightforward way to represent this variable.

For the "Education Level" variable, you would use one-hot encoding, which assigns a unique binary vector 
to each category. This allows the model to distinguish between different levels of education.

For the "Employment Status" variable, you would use ordinal encoding, which assigns a numerical value to 
each category. This allows the model to understand the relative importance of each category.

Finally, you could also use dummy encoding, which assigns a binary vector to each category. This is useful
if you want to represent the categories as independent variables.

"""Q7. You are analyzing a dataset with two continuous variables, "Temperature" and "Humidity", and two
categorical variables, "Weather Condition" (Sunny/Cloudy/Rainy) and "Wind Direction" (North/South/
East/West). Calculate the covariance between each pair of variables and interpret the results."""

Ans: Covariance between Temperature and Humidity: 
This is a measure of how much the two variables vary together. A positive covariance indicates that as one 
variable increases, the other also tends to increase; a negative covariance indicates that as one variable
increases, the other tends to decrease.

Covariance between Weather Condition and Wind Direction: 
This is a measure of how much the two variables are related. A positive covariance indicates that when one 
variable increases, the other tends to increase as well; a negative covariance indicates that when one 
variable increases, the other tends to decrease.

Covariance between Temperature and Weather Condition: 
This is a measure of how much the two variables vary together. A positive covariance indicates that as 
one variable increases, the other also tends to increase; a negative covariance indicates that as one 
variable increases, the other tends to decrease.

Covariance between Temperature and Wind Direction: 
This is a measure of how much the two variables vary together. A positive covariance indicates that as 
one variable increases, the other also tends to increase; a negative covariance indicates that as one 
variable increases, the other tends to decrease.
```
